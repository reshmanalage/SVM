```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

# Drop the 'Unnamed: 0' column as it seems unnecessary
data = data.drop(columns=['Unnamed: 0'])


```

```python
# Separate features (X) and target variable (y)
X = data.drop(columns=['class'])
y = data['class']

# One-hot encode categorical variables
encoder = OneHotEncoder(drop='first')
X_encoded = encoder.fit_transform(X)

# Split the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)

# Check the shapes of the resulting datasets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
import numpy as np
import scipy.stats as ss

# Define a function to plot histograms and count plots
def plot_distribution(feature):
    plt.figure(figsize=(10, 6))
    if data[feature].dtype == 'float64':
        sns.histplot(data[feature], kde=True)
    else:
        sns.countplot(x=data[feature].astype('category'))
    plt.title(f"Distribution of {feature}")
    plt.show()

# Plot histograms and count plots for each feature
for column in data.columns:
    plot_distribution(column)
```

```python
# Compute the correlation matrix using Cramer's V
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k - 1) * (r - 1)) / (n - 1))
    rcorr = r - ((r - 1) ** 2) / (n - 1)
    kcorr = k - ((k - 1) ** 2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

# Compute the Cramer's V correlation matrix for categorical features
categorical_features = data.select_dtypes(include='object').columns
corr_matrix = pd.DataFrame(index=categorical_features, columns=categorical_features)
for i in range(len(categorical_features)):
    for j in range(len(categorical_features)):
        corr_matrix.iloc[i, j] = cramers_v(data[categorical_features[i]], data[categorical_features[j]])

# Plot the correlation matrix
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix.astype(float), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Cramer's V Correlation Matrix")
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Implement a basic SVM classifier
svm_model = SVC(kernel='linear', random_state=42)

# Train the SVM model on the training data
svm_model.fit(X_train, y_train)

# Predict on the test data
y_pred = svm_model.predict(X_test)

# Evaluate model performance
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)
```

```python
# Select numeric features only
selected_features = ['stalk_height', 'cap_diameter']
X_numeric = data[selected_features]
y = data['class']

# Split the numeric data into training and testing sets
X_train_num, X_test_num, y_train, y_test = train_test_split(X_numeric, y, test_size=0.2, random_state=42)

# Encode target variable
y_train = y_train.map({'edible': 0, 'poisonous': 1})
y_test = y_test.map({'edible': 0, 'poisonous': 1})
```

```python
# Train the SVM model
svm_model_num = SVC(kernel='linear', C=1)
svm_model_num.fit(X_train_num, y_train)

# Function to plot decision boundary for numeric features
def plot_decision_boundary(X, y, model, feature_names):
    x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
    y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.figure(figsize=(10, 6))
    plt.contourf(xx, yy, Z, alpha=0.3)
    scatter = plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, edgecolors='k', marker='o')
    plt.xlabel(feature_names[0])
    plt.ylabel(feature_names[1])
    plt.title('Decision Boundary')
    plt.legend(handles=scatter.legend_elements()[0], labels=['edible', 'poisonous'])
    plt.show()

# Plot decision boundary
plot_decision_boundary(X_train_num, y_train, svm_model_num, selected_features)
```

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'gamma': ['scale', 'auto']
}

# Grid search for hyperparameter tuning
grid_search = GridSearchCV(SVC(), param_grid, refit=True, verbose=3)
grid_search.fit(X_train, y_train)

# Best parameters and model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

print(f"Best Parameters: {best_params}")

# Evaluate the best model
y_pred_best = best_model.predict(X_test)
best_accuracy = accuracy_score(y_test, y_pred_best)
best_conf_matrix = confusion_matrix(y_test, y_pred_best)
best_class_report = classification_report(y_test, y_pred_best)

print(f"Best Model Accuracy: {best_accuracy}")
print("Best Model Confusion Matrix:")
print(best_conf_matrix)
print("Best Model Classification Report:")
print(best_class_report)
```

```python
# Define parameter grid with class weights
param_grid_with_weights = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto'],
    'class_weight': ['balanced']
}

# Perform grid search with class weights
grid_search_with_weights = GridSearchCV(SVC(), param_grid_with_weights, refit=True, verbose=3, cv=5)
grid_search_with_weights.fit(X_train, y_train)

# Best parameters and model from grid search with class weights
best_params_with_weights = grid_search_with_weights.best_params_
best_model_with_weights = grid_search_with_weights.best_estimator_

print(f"Best Parameters with Class Weights: {best_params_with_weights}")

# Evaluate the best model with class weights
y_pred_best_with_weights = best_model_with_weights.predict(X_test)
best_accuracy_with_weights = accuracy_score(y_test, y_pred_best_with_weights)
best_conf_matrix_with_weights = confusion_matrix(y_test, y_pred_best_with_weights)
best_class_report_with_weights = classification_report(y_test, y_pred_best_with_weights)

print(f"Best Model Accuracy with Class Weights: {best_accuracy_with_weights}")
print("Best Model Confusion Matrix with Class Weights:")
print(best_conf_matrix_with_weights)
print("Best Model Classification Report with Class Weights:")
print(best_class_report_with_weights)
```
